# Give id to companies and create company metadata

```python
companynames=list(test['CompanyName'].unique())
Col_names=['CompanyName','ID','State','City','NAICS']
companytoid=pd.DataFrame(columns=Col_names)

for i in range(len(companynames)):
    row=test.loc[test["CompanyName"]==companynames[i]].head(1)
    #print(row)
    state=row['State'].values[0]
    city=row['City'].values[0]
    x=row['NAICS'].values[0]
    code=int(str(x)[:2])
    
    companytoid=companytoid.append({'CompanyName':companynames[i], 'ID':i, 'State':state,'City':city,'NAICS':code}, ignore_index = True)
    
    train.loc[train["CompanyName"]==companynames[i],"Company_id"]=i
    test.loc[test["CompanyName"]==companynames[i],"Company_id"]=i
    val.loc[val["CompanyName"]==companynames[i],"Company_id"]=i
    
    
companytoid.to_pickle("/data/WorkData/firmEmbeddings/JoinDataPrep/Company_meta.pkl")
companytoid.head(5)
```
